```python
import numpy as np
def possible_machine(graph,order,x,y,job,machine):
    index_1 = graph[machine-1][:].index(job)
    for i in range(0,len(x)):
        if y[i] == machine:         
            index_2 = graph[machine-1][:].index(x[i])
            if index_1 < index_2:
                return False
    return True
def insert_job(graph,order,x,y,job,machine,possible):
    index = 0
    if not possible:
        for i in range(0,len(x)):
            if y[i] == machine:
                index_1 = graph[machine-1][:].index(job)
                index_2 = graph[machine-1][:].index(x[i])
                if index_1 < index_2:
                    break
            index += 1
    else:
        for i in range(0,len(x)):
            if x[i] == job:
                index_1 = order[job-1][:].index(machine)
                index_2 = order[job-1][:].index(y[i])
                if index_1 < index_2:
                    break
            index += 1
    if index == len(x):
        x.append(job)
        y.append(machine)
    else:
        x.insert(index,job)
        y.insert(index,machine)
    return index
def fix(graph,x,y,job,machine,index,work):
    index_1 = graph[machine-1][:].index(job)
    for i in range(0,index):
        if y[i] == machine:         
            index_2 = graph[machine-1][:].index(x[i])
            if index_1 < index_2:
                y[i] = -1
                x[i] = -1
                work[machine-1][index_2] = 0
    while -1 in x or -1 in y:
        x.remove(-1)
        y.remove(-1)
def graph_to_individual(graph,order,j,m):
    jobs = []
    machines = []
    work = np.zeros((m,j),dtype=int)
    while 0 in work:
        for mach in range(0,m):
            for jb in range(0,j):
                if not work[mach][jb]:
                    possible = possible_machine(graph,order,jobs,machines,graph[mach][jb],mach+1)
                    print(possible,graph[mach][jb],mach+1)
                    index = insert_job(graph,order,jobs,machines,graph[mach][jb],mach+1,possible)
                    print(jobs)
                    print(machines)
                    if index == -1:
                        index = len(jobs)-1
                    fix(graph,jobs,machines,graph[mach][jb],mach+1,index,work)
                    print(jobs)
                    print(machines)
                    print("-------")
                    work[mach][jb] = 1
        print(work)
    return jobs,machines
graph = [[1,2],
         [1,2]]
order = [[1,2],
         [2,1]]
j = 2
m = 2
```
